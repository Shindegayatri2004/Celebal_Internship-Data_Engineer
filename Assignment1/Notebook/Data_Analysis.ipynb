#Step 4: Data Cleaning

```python
numeric_cols = ['rating',
                'ratings_count',
                'initial_price',
                'discount',
                'final_price']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print(df.isnull().sum())
```

```python
for col in numeric_cols:
    if not df[col].isnull().all():
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value)
    else:
        print(f"Skipping {col} because it contains only NaN values.")
```

```python
print("Duplicate Rows :", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("Shape after removing duplicates :", df.shape)
```

```python
selected_columns = df[['title',
                       'rating',
                       'final_price',
                       'category']]

selected_columns.head()
```

```python
high_rating = df[df['rating'] > 4]

high_rating.head()
```

```python
expensive_products = df[df['final_price'] > 1000]

expensive_products.head()
```

```python
df['price_difference'] = df['initial_price'] - df['final_price']
```

```python
df['quantity'] = np.random.randint(1,10,size=len(df))
```

```python
df['total_amount'] = df['final_price'] * df['quantity']
```

```python
df['popularity_metric'] = (
    df['rating'] * np.log1p(df['ratings_count'])
)
```

```python
df['discount_percent'] = (
    (df['initial_price']-df['final_price'])
    / df['initial_price']
)*100
```

```python
category_count = df['category'].value_counts()

print(category_count)
```

```python
avg_price = df.groupby('category')['final_price'].mean()

print(avg_price.sort_values(ascending=False))
```

```python
avg_rating = df.groupby('category')['rating'].mean()

print(avg_rating.sort_values(ascending=False))
```

```python
corr = df[['rating',
           'ratings_count',
           'initial_price',
           'final_price',
           'price_difference',
           'total_amount',
           'popularity_metric']]

corr.corr()
```

##Visualization

```python
#Rating Distribution
plt.figure(figsize=(8,5))

sns.histplot(df['rating'], bins=20)

plt.title("Distribution of Ratings")

plt.savefig("rating_distribution.png")

plt.show()
```

```python
#Category Distribution
plt.figure(figsize=(12,6))

df['category'].value_counts().head(10).plot(kind='bar')

plt.title("Top Categories")

plt.xticks(rotation=45)

plt.savefig("category_distribution.png")

plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10,8))

sns.heatmap(corr.corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.savefig("correlation_heatmap.png")

plt.show()
```

```python
#Rating vs Popularity
plt.figure(figsize=(8,6))

sns.scatterplot(x='rating',
                y='popularity_metric',
                data=df)

plt.title("Rating vs Popularity")

plt.savefig("popularity_vs_rating.png")

plt.show()
```
